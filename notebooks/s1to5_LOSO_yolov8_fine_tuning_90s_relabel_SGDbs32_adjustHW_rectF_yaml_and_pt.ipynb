## PT

```python
from ultralytics import YOLO

# Hyperparameters
EPOCHS = 100
BATCH_SIZE = 32
IMG_SIZE = (640, 1280)
LEARNING_RATE = 0.001
MOMENTUM = 0.9
WEIGHT_DECAY = 0.0002
WARMUP_EPOCHS = 3
WARMUP_MOMENTUM = 0.8
WARMUP_BIAS_LR = 0.1
OPTIMIZER = "SGD"
AUGMENT = True
PATIENCE = 10

def main():
    # Pass the YAML file path as a string
    yaml_path = "yolov8n.pt"  # This should remain a path, not the loaded contents

    # Load the YOLO model with the YAML file path
    model = YOLO(yaml_path)  # Use the YAML file path as a string

    # Train the model with the specified hyperparameters
    model.train(
        data="re_loso_fold_s01_90s.yaml",
        epochs=EPOCHS,
        batch=BATCH_SIZE,
        imgsz=IMG_SIZE,
        lr0=LEARNING_RATE,
        momentum=MOMENTUM,
        weight_decay=WEIGHT_DECAY,
        warmup_epochs=WARMUP_EPOCHS,
        warmup_momentum=WARMUP_MOMENTUM,
        warmup_bias_lr=WARMUP_BIAS_LR,
        optimizer=OPTIMIZER,
        verbose=False,  # Set this to False to suppress most logging
        augment=AUGMENT,
        patience=PATIENCE,
        name="LOSO_re_kfold_s01_90s_rightHW_rectF_bs32_SGD_pt"
    )

if __name__ == "__main__":
    main()
```
